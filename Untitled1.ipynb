```python
def generate_unique_responses(input_ids, num_samples=20):
    sample_outputs = model.generate(input_ids, 
                                    num_return_sequences=num_samples,
                                    do_sample=True, 
                                    max_length=50, 
                                    top_k=10, 
                                    top_p=0.95,
                                   temperature = 1.2)

    # Process all outputs at once
    decoded_outputs = [tokenizer.decode(output, skip_special_tokens=True) for output in sample_outputs]
    return set(decoded_outputs)

# Tokenize all inputs at once
tokenized_inputs = [tokenizer.encode(query, return_tensors='pt') for query in df_test['query']]
acc = []

for index, (input_ids, expected_output) in enumerate(zip(tokenized_inputs, df_test['doc_id'])):
    unique_responses = generate_unique_responses(input_ids)

    if expected_output in unique_responses:
        acc.append(1)
    else:
        acc.append(0)

# Compute accuracy
accuracy = sum(acc) / len(acc)
```

### Concatenating all personal_dfs

```python
datasets_folder = os.path.join('data', 'datasets')
df_train = pd.read_csv(os.path.join('data', 'datasets', 'train_df.csv'))
df_test = pd.read_csv(os.path.join('data', 'datasets', 'test_df.csv'))
personal_dfs = os.listdir(datasets_folder)
personal_dfs = [i for i in personal_dfs if '.csv' in i and 'train' not in i and 'test' not in i]
df_total = pd.DataFrame()
for i in personal_dfs:
    df_temp = pd.read_csv(os.path.join(datasets_folder, i))
    df_total = pd.concat([df_total, df_temp])
    
df_total = df_total.drop_duplicates()

df_test = df_test[df_test['doc_id'].isin(df_total['doc_id'].unique())]

```

```python
def read_model_and_evaluate(model_name):
    df_tot = df_total.copy()
    df_tst = df_test.copy()
    model = T5ForConditionalGeneration.from_pretrained(os.path.join(models_folder, model_name))
    tokenizer = T5Tokenizer.from_pretrained('t5-small')
    df_tot['generated_doc_id'] = df_tot['query'].apply(lambda x: generate_text(x, model))
    df_tst['generated_doc_id'] = df_tst['query'].apply(lambda x: generate_text(x, model))

    print ('MODEL:', model_name)
    print ('training set accuracy: ', 
            df_tot[df_tot['doc_id'] == df_tot['generated_doc_id']].shape[0]/df_tot.shape[0])
    print ('test set accuracy: ',
            df_tst[df_tst['doc_id'] == df_tst['generated_doc_id']].shape[0]/df_tst.shape[0])
```

```python
model_name = '8098_2023-12-06 055305_my_t5_model'
read_model_and_evaluate(model_name)

model_name = '8099_2023-12-06 055346_my_t5_model'
read_model_and_evaluate(model_name)


print ('PAST VERSIONS OF MODEL 8099:')
model_name = '8099_2023-12-06 033709_my_t5_model'
read_model_and_evaluate(model_name)

model_name = '8099_2023-12-06 011939_my_t5_model'
read_model_and_evaluate(model_name)
```
